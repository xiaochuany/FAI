```python
import os
import numpy as np
import pandas as pd
```

```python
csv_directory = "./data"

csv_files = [f for f in os.listdir(csv_directory) if f.endswith('.csv')]

dataframes = {}

for csv_file in csv_files:
    csv_path = os.path.join(csv_directory, csv_file)
    df = pd.read_csv(csv_path)
    name = os.path.splitext(csv_file)[0]
    dataframes[name]=df
```
